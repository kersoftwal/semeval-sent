```python
import torch
import torch.nn as nn

class FirstNeuralNetwork(nn.Module):
    
    def __init__(self, input_size, hidden_size, num_classes):
        super().__init__()
        self.layer1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(hidden_size, num_classes)
        
    def forward(self, data_input):
        output_layer1 = self.layer1(data_input)
        output_layer1 = self.relu(output_layer1)
        output_layer2 = self.layer2(output_layer1)
        return output_layer2

first_neural_network = FirstNeuralNetwork(300, 100, 2)
print(first_neural_network)
        
```

```python
import numpy as np

# substituting words with ids 
word_map = {word:idx for idx,word in enumerate(model.wv.vocab, start=2)}
training_sentence_data = (
    text.apply(lambda x: list(map(lambda word: word_map[word] if word in word_map else 1, x.split(' '))))
)

# padding zeros
training_sentence_data = (
    list(map(lambda sent: np.pad(sent,(0,sentence_length-len(sent))), training_sentence_data))
)

# from numpy arrays to tensors
training_sentence_data = (
    list(map(lambda sent_np: torch.LongTensor(sent_np), training_sentence_data))
)

training_sentence_data[0].shape
```

```python
np.random.seed(100)
model.wv.vectors.shape
```

```python
word_vectors_for_training = np.insert(
    model.wv.vectors,
    0,
    np.random.uniform(model.wv.vectors.min(), model.wv.vectors.max(), 100),
    axis=0
)

word_vectors_for_training = np.insert(
    word_vectors_for_training,   
    0, 
    np.zeros(100),
    axis=0
)

word_vectors_for_training = torch.FloatTensor(word_vectors_for_training)
word_vectors_for_training.shape
```

```python
class FirstNeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super().__init__()
        self.embedding_layer = nn.EmbeddingBag.from_pretrained(word_vectors_for_training, mode='mean')
        self.layer1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.layer2 = nn.Linear(hidden_size, num_classes)
        
    def forward(self, data_input):
        embedded_data_input = self.embedding_layer(data_input)
        output_layer1 = self.layer1(embedded_data_input)
        output_layer1 = self.relu(output_layer1)
        output_layer2 = self.layer2(output_layer1)
        return output_layer2.squeeze()

training_sentence_data = torch.stack(training_sentence_data)
first_neural_network = FirstNeuralNetwork(100, 50, 4)
first_neural_network
```
